```python
# Generate simulated temperature readings from two sensors.
# Sensor A: mean 25 C, std 3 C, 200 readings.
# Sensor B: mean 27 C, std 4.5 C, 200 readings.
# Also generate 200 timestamps uniformly from 0 to 10 seconds.
# Use np.random.default_rng with a seed = last 4 digits of your Drexel ID.

import numpy as np

# Set the random seed using the last 4 digits of Drexel ID
seed = 2682  # Last 4 digits of Drexel ID
rng = np.random.default_rng(seed=seed)

# Generate 200 timestamps uniformly distributed from 0 to 10 seconds
timestamps = rng.uniform(0, 10, 200)

# Generate temperature readings for Sensor A: normal distribution, mean 25°C, std 3°C
sensor_a_temps = rng.normal(25, 3, 200)

# Generate temperature readings for Sensor B: normal distribution, mean 27°C, std 4.5°C
sensor_b_temps = rng.normal(27, 4.5, 200)
```

```python
# Scatter plot: sensor reading (y-axis) vs timestamp in seconds (x-axis).
# Color Sensor A points blue and Sensor B points orange.
# Add axis labels with units, a title, and a legend.

import matplotlib.pyplot as plt

plt.scatter(timestamps, sensor_a_temps, color='blue', label='Sensor A')
plt.scatter(timestamps, sensor_b_temps, color='orange', label='Sensor B')
plt.xlabel('Timestamp (seconds)')
plt.ylabel('Sensor Reading (°C)')
plt.title('Sensor Readings vs Time')
plt.legend()
plt.show()
```

```python
# Overlaid histogram of Sensor A and Sensor B temperature distributions.
# Use 30 bins, alpha=0.5 for transparency so both distributions are visible.
# Add vertical dashed lines at each sensor's mean.
# Include a legend labeling each sensor.

import matplotlib.pyplot as plt

plt.hist(sensor_a_temps, bins=30, alpha=0.5, label='Sensor A', color='blue')
plt.hist(sensor_b_temps, bins=30, alpha=0.5, label='Sensor B', color='orange')
plt.axvline(sensor_a_temps.mean(), color='blue', linestyle='--', label='Sensor A Mean')
plt.axvline(sensor_b_temps.mean(), color='orange', linestyle='--', label='Sensor B Mean')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.title('Temperature Distributions of Sensors A and B')
plt.legend()
plt.show()
```

```python
# Side-by-side box plot comparing Sensor A and Sensor B distributions.
# Label x-axis with sensor names, y-axis with "Temperature (deg C)".
# Add a horizontal dashed line at the overall mean of both sensors combined.

import matplotlib.pyplot as plt
import numpy as np

# Calculate the overall mean of both sensors combined
overall_mean = np.concatenate([sensor_a_temps, sensor_b_temps]).mean()

# Create side-by-side box plot
plt.boxplot([sensor_a_temps, sensor_b_temps], labels=['Sensor A', 'Sensor B'])
plt.axhline(overall_mean, color='red', linestyle='--', label=f'Overall Mean: {overall_mean:.2f}°C')
plt.ylabel('Temperature (deg C)')
plt.title('Box Plot Comparison of Sensor A and B Temperatures')
plt.legend()
plt.show()
```
